```python
%matplotlib inline

import bloscpack as bp
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import math
import pandas as pd
import os
import sys 

sys.path.append(os.path.join(os.environ['REPO_DIR'], 'utilities'))
from metadata import *
from utilities2015 import *
from registration_utilities import *
from annotation_utilities import *
from data_manager import DataManager



stack = 'DK37'

output_dir = os.path.join(ROOT_DIR, 'CSHL_data_processed', stack, 'prep1_raw_NtbNormalized_manual/')

# Input specs (resolution will carry over)
resolution = 'thumbnail'
prep_id = 'None'
version = 'NtbNormalized'
```

```python
resolution = 'raw'
prep_id = 'None'
version = 'Ntb'
```

```python
# Anchor transform
#for section in range(150, 155):
for section in metadata_cache['valid_sections_all'][stack]:
    fn = metadata_cache['sections_to_filenames'][stack][section]
    
    if fn=='Placeholder':
        continue
        
    img_fp = DataManager.get_image_filepath_v2(stack=stack, section=section, 
                            prep_id=prep_id, resol=resolution, version=version)
    
    if resolution=='thumbnail':
        T = DataManager.load_transforms(stack, downsample_factor=32, use_inverse=True)[fn]
    elif resolution=='raw':
        T = DataManager.load_transforms(stack, downsample_factor=1, use_inverse=True)[fn]
    else:
        print('Bad resolution')
        break
    
    op_str = ''
    op_str += " +distort AffineProjection '%(sx)f,%(rx)f,%(ry)f,%(sy)f,%(tx)f,%(ty)f' " % {
     'sx':T[0,0],
     'sy':T[1,1],
     'rx':T[1,0],
     'ry':T[0,1],
     'tx':T[0,2],
     'ty':T[1,2],}
    
    if resolution=='thumbnail':
        x = 0
        y = 0
        w = 2001
        h = 1001
    else:
        x = 0
        y = 0
        w = 64001
        h = 32001
    op_str += ' -crop %(w)sx%(h)s%(x)s%(y)s\! ' % {
                        'x': '+' + str(x) if int(x) >= 0 else str(x),
                        'y': '+' + str(y) if int(y) >= 0 else str(y),
                        'w': str(w),
                        'h': str(h)}
    
#     output_fp ='/home/alexn/Desktop/transforms/'
    output_fp = output_dir
    if not os.path.exists( output_fp ):
        os.makedirs(output_fp)

    try:
        execute_command("convert \"%(input_fp)s\"  -normalize -depth 8 +repage -virtual-pixel background -background %(bg_color)s %(op_str)s -flatten -compress lzw \"%(output_fp)s\"" % \
                    {'op_str': op_str,
                     'input_fp': img_fp,
                     'output_fp': output_fp+fn+'.tif',
                     'bg_color': 'black'})
    except Exception as e:
        print e
```
